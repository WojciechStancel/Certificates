```python
import numpy as np
import pandas as pd
```

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

```python
df.to_csv('meta.csv')
```

```python
df_nov = df[(df.index.month == 2) & (df.index.year == 2023)]
df_nov
```

```python
df_nov.to_csv('fb_nov.csv')
```

```python
new_df = pd.read_csv('fb_nov.csv')
new_df
```

```python
new_df = pd.read_csv('fb_nov.csv', index_col=0)
new_df
```
