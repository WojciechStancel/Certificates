```python
df['Average'] = (df.Open + df.Close) / 2.0
df.head()
```

```python
df.sort_index()
```

```python
df = df.sort_index()
df.head()
```

```python
df[['Close']].shift(1)
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head()
```

```python
df['Daily_Change'] = df.Close / df.Close_shift - 1
df.head()
```
